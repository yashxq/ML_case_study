## 2. Hypothesis Testing

**2.a. 2- Sample T-Test to check if Working Day has an effect on the number of electric cycles rented (10 points)**

ho: count on working day < count on non-working day\
ha: count on working day >= count on non-working day

```python
workingday_sample = df["workingday"].sample(n=1000)
count_sample = df["count"].sample(n=1000)

stats.ttest_ind(workingday_sample, count_sample, alternative="less")
```

Inference: pvalue<<0.05, we reject our Null hypothesis and accept ha.

Therefore, count on working day >= count on non-working day

**2.b. ANNOVA to check if No. of cycles rented is similar or different in different 1. weather 2. season (10 points)**

**2.b.1. weather**

ho: count is same in different weathers \
h1: count is not same in different weathers 

```python
df.weather.value_counts()
```

```python
df.groupby('weather').agg({'count':'sum'}).reset_index(drop=True)
```

```python
weather_1 = df[df["weather"]==1]["count"].sample(n=500)
weather_2 = df[df["weather"]==2]["count"].sample(n=500)
weather_3 = df[df["weather"]==3]["count"].sample(n=500)

weather_1 = weather_1.reset_index(drop=True)
weather_2 = weather_2.reset_index(drop=True)
weather_3 = weather_3.reset_index(drop=True)

dataset = pd.DataFrame({"1":weather_1, "2":weather_2, "3":weather_3})
dataset
```

```python
f, pval = stats.f_oneway(dataset["1"], dataset["2"], dataset["3"])
print(f, pval)
```

Inference: Since pvalue<0.05, we reject ho and go with ha.

```python
mean_1 = dataset['1'].mean()
mean_2 = dataset['2'].mean()
mean_3 = dataset['3'].mean()

print(mean_1, mean_2, mean_3)
```

```python
sns.distplot(dataset['1'], bins = 10)
```

```python
sm.qqplot(dataset['1'], line = "s")
py.show()
```

Inference: weather_1 is not following Normal distribution.

Lets try to do Bootstrapping and see if Sampling distribution follow Normal distribution or not. Irrespective of how your distribution is, if your smapling distribution follows a Normal distribution, we can approximate from CLT.

```python
weather_1_samp = []
for i in range(500):
    w1 = df[df["weather"]==1]["count"].sample(n=200)
    av = w1.mean()
    weather_1_samp.append(av)
```

```python
sns.distplot(weather_1_samp, bins = 20)
```

Inference: This is following Normal Distribution.

```python
np.mean(weather_1_samp)
```

```python
mu = (mean_1*500 + mean_2*500 + mean_3*500)/1500
mu
```

```python
ssw = sum((weather_1-mean_1)**2 + (weather_2-mean_2)**2 + (weather_3-mean_3)**2)
ssw
```

```python
ssg = ((mean_1 - mu)**2)*500 + ((mean_2 - mu)**2)*500 + ((mean_3 - mu)**2)*500
ssg
```

```python
msg = ssg/2
msg
```

```python
msw = ssw/(1500-3)
msw
```

```python
fstat = msg/msw
fstat
```

Then to get Fcritical values look at the F-distribution table.



fcrit~=3.0

Since fcrit<<45.198, we reject ho and accept ha.

**2.c. Chi-square test to check if Weather is dependent on the season (10 points)**

Hypothesis Testing:\
ho: There is no relationship btw weather and season.\
ha: There is relationship btw weather and season.

```python
ctab = pd.crosstab(df['weather'], df['season'])
ctab
```

```python
from scipy.stats import chi2_contingency
```

```python
stat, p, dof, expected = chi2_contingency(ctab)
```

```python
print("Chi-square stat:",stat)
print("pvalue :",p)
```

dof = (r-1)*(c-1) = 9

alpha = 0.05

chi-stats = 49.95

See the Chi-square distribution table to know the chi-crit.\
chi-crit = 16.919

chi-crit < chi-stats, so we reject ho and accepts ha.

Therefore, there is relationship btw weather and season ie weather and season are dependent on each other.
