```python
trip['destination_name'] = trip['destination_name'].str.lower()
trip['source_name'] = trip['source_name'].str.lower()
```

```python
def get_state(x):
    if x==None:
        return x
    
    state = x.split('(')[1]
    return state[:-1]

def get_city(x):
    if x==None:
        return x
    city = x.split(' (')[0]
    
    city = city.split('_')[0]
    
    if city =='pnq vadgaon sheri dcp':
        return 'vadgaonsheri'
    if city in ['pnq pashan dcp','pnq rahatani dcp','pune balaji nagar']:
        return 'pune'
    if city == 'hbr layout pc':
        return 'bengaluru'
    if city == 'bhopal mp nagar':
        return 'bhopal'
    if city == 'mumbai antop hill':
        return 'mumbai'
    
    return city

def get_place(x):
    if x==None:
        return x
    x = x.split(" (")[0]
    
    total_len = len(x.split('_'))
    
    if total_len>=3:
        return x.split('_')[1]
    if total_len ==2:
        return x.split('_')[0]
    
    
    return x.split(' ')[0]
def get_code(x):
    if x==None:
        return x
    x= x.split(' (')[0]
    if len(x.split('_')) >=3:
        return x.split('_')[-1]
    return 'none'
```

```python
trip['destination_state'] = trip['destination_name'].apply(lambda x :get_state(x))
trip['destination_city'] = trip['destination_name'].apply(lambda x :get_city(x))
trip['destination_place'] = trip['destination_name'].apply(lambda x :get_place(x))
trip['destination_code'] = trip['destination_name'].apply(lambda x :get_code(x))
```
